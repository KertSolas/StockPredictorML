```python
import pandas as pd
import yfinance as yf
from datetime import date
from prophet import Prophet
from prophet.plot import plot_plotly
from plotly import graph_objs as go
```

```python
today = date.today()
# startYear = today.year
# startYear = 0
# endYear = startYear + 5
year = 0
period = 0 * 365
```

```python
def getStockData(symbol, start_date, end_date):
    data = pd.DataFrame(yf.download(symbol, start=start_date, end=end_date))
    data.reset_index(inplace=True)
    return data
```

```python
data = getStockData('AAPL', '2012-01-01', '2021-01-01')
```

```python
df_train = data[["Date", "Close", "Volume"]]
df_train = data.rename(columns={"Date": "ds", "Close": "y"})
model = Prophet()
model.fit(df_train) 
```

```python
future = model.make_future_dataframe(periods= period)
forecast = model.predict(future)
forecast
```
